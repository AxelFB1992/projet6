## Configuration de l'environnement de travail

### Creation de l'environnement virtuel

Ci-dessous les commandes que j'ai utilisé pour créer, configurer et executer mon environnement virtuel :

- python3 -m venv jupyterlab-env : creer une environnement virtuel python grâce à l'outil venv que l'on nommera jupyterlab-env. C'est venv qui crée l'envrionnement mais il a besoin de l'environnement python3 pour s'executer.
- source jupyterlab-env/bin/activate : Executer l'environnement virtuel jupyterlab-env.
- pip install jupyterlab : Installer jupyterlab si il n'a pas été inscrit dans le plan de montage de l'environnement virtuel (requirements.txt).
- pip freeze > requirements.txt : Cela permettra d'installer automatiquement jupyterlab lorsqu'on re-installera un envrionnement virtuel à partir de ce requirements.txt.
- jupyter lab : lancer le serveur jupyterlab sous l'environnement virtuel jupyterlab-env (en dehors, cela ne marchera pas, à moins que jupyterlab soit installé en dur sur la machine). Il faut ensuite copier/coller dans un navigateur l'url qui nous est donné contenant un token. Ici en l'occurence, il s'agit de http://localhost:8888/?token=8bffb0ff13b4c159e1f90ebe7d6922ff992a9f6d503c545e.
